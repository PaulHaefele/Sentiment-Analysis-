## Text Preprocessing: Lemmatization

In the context of this sentiment analysis project, we have chosen to apply lemmatization as a text preprocessing technique to prepare the 'reviewText' column. Lemmatization is a linguistically motivated method that reduces words to their base or dictionary form, known as a lemma. It is a valuable technique for sentiment analysis for several reasons:

1. **Contextual Accuracy**: Lemmatization provides a more contextually accurate representation of words in the text. It transforms words into their base forms while considering their grammatical meanings. This ensures that the resulting words remain valid and meaningful.

2. **Retains Word Integrity**: Unlike stemming, which might truncate words to their root form, lemmatization retains the integrity of words. This is especially important in sentiment analysis, as it helps preserve the meaning of words and nuances that may influence sentiment.

3. **Improved Semantic Understanding**: By lemmatizing the text, we can better capture the semantic nuances present in reviews. For example, lemmatization ensures that words like "running" and "ran" are both reduced to "run," preserving the underlying semantic connection.

4. **High-Quality Analysis**: When conducting sentiment analysis, the accuracy of text preprocessing directly impacts the quality of the analysis results. Lemmatization, due to its context-aware nature, enhances the quality of the text data before sentiment analysis begins.

We believe that by applying lemmatization to the 'reviewText' column, we can prepare the data in a way that reflects the richness of language used in the reviews and ultimately improve the accuracy of our sentiment analysis. It's an essential step in ensuring that we extract meaningful sentiment insights from the textual data.

Please note that we will carefully document the lemmatization process, and any subsequent data wrangling steps, to maintain transparency and repeatability in our analysis.


```python
from nltk.stem import WordNetLemmatizer
 
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = text.split()  # Tokenize the text into words
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)  # Rejoin the lemmatized words

# Apply lemmatization to the entire 'reviewText' column
df['reviewText'] = df['reviewText'].apply(lemmatize_text)
```

## Preserving Context with Stop Words

In this analysis, we have decided to retain stop words in our text data. Stop words are common words such as "the," "and," "is," and "not" that are typically removed during text preprocessing. The primary reason for this decision is to preserve the contextual information present in the reviews.

**Why Keep Stop Words?**

1. **Negation Handling**: Stop words like "not" play a crucial role in negating sentiment. For instance, consider the review that starts with "No issues." The presence of "not" in this context is essential to understanding the sentiment expressed.

2. **Intensity and Modifiers**: Some stop words, like "very" or "extremely," modify the intensity of sentiment words. Removing these stop words would result in a loss of intensity and nuanced sentiment expression.

3. **Contextual Understanding**: Preserving stop words helps us maintain a better understanding of the context in which sentiment words are used. This contextual information is vital for accurate sentiment analysis.

4. **Idiomatic Expressions**: Stop words are sometimes part of idiomatic expressions or common phrases that carry specific sentiment. Removing these stop words could lead to the loss of these meaningful phrases.

**An Example**

Consider the review, "No issues and 4 stars." In this review, the presence of "No issues" is a clear example of the importance of preserving stop words. The combination of "No" and "issues" conveys a specific sentiment that would be lost if stop words were removed. 

By keeping stop words, we aim to maintain the richness and context of the reviews, ensuring that our sentiment analysis accurately captures the intended sentiment expressions.

We will proceed with this approach to maintain the integrity of the text data for our sentiment analysis.

```python
# Plot the distribution of user ratings
plt.figure(figsize=(8, 6))
plt.hist(df['userRating'], bins=5, edgecolor='k', alpha=0.7)
plt.title('Distribution of User Ratings')
plt.xlabel('User Rating')
plt.ylabel('Count')
plt.xticks(range(1, 6))  # Set x-axis ticks for ratings from 1 to 5
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
from wordcloud import WordCloud

# Define a function to generate the word cloud without stop words
def generate_word_cloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=None).generate(text)
    return wordcloud

# Combine all the review text into a single string
all_text = ' '.join(df['reviewText'])

# Create the word cloud without stop words
wordcloud = generate_word_cloud(all_text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Frequency Analysis (Excluding Stop Words)')
plt.show()
```
